```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import math
```

```python
signal=np.random.randint(0,2,20)
print(signal)
```

```python
a1=[]
for i in range(0,len(signal),2):
    a1.append(signal[i])

print(a1)
```

```python
a2=[]
for i in range(1,len(signal),2):
    a2.append(signal[i])

print(a2)
```

```python
fc = 1000
T= 1/fc
t=np.arange(-10/fc , 25/fc , 0.01/fc)
```

```python
def p1(t):     
    return np.sqrt(2/T)*np.cos(2 * (np.pi) * fc * t) * (abs(t) <= T)
```

```python
def p2(t):     
    return np.sqrt(2/T)*np.sin(2 * (np.pi) * fc * t) * (abs(t) <= T)
```

```python
for i in range(len(a1)):
    c=a1[i]*p1(t-(2/fc)*i) + a2[i]*p2(t-(2/fc)*i)
    plt.plot(t,c)
```
